Pada gambar `tomat_jeruk.jpg`, buatlah sebuah filter untuk menangkap tomat-nya saja. Tentunya akan lebih mudah kalau kamu membuka dengan colorspace HSV.

```python
img_tomat_bgr = cv2.imread("tomat_jeruk.jpg", cv2.IMREAD_COLOR)
print("img_tomat_bgr shape is: ", img_tomat_bgr.shape)

plt.imshow(img_tomat_bgr[:,:,::-1])
```

```python
hsv_tomat = cv2.cvtColor(img_tomat_bgr,cv2.COLOR_BGR2HSV)
```

```python
hsv_tomat.shape
```

```python
kuning_bawah = np.array([100,0,0])
kuning_atas  = np.array([150,255,255])
mask1 = cv2.inRange(hsv_tomat,kuning_bawah,kuning_atas)
```

```python
plt.imshow(mask1, cmap='gray')
```
